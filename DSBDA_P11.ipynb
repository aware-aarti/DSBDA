```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna().isnull().sum()
```

Use the dataset 'titanic'. Plot a box plot for distribution of age with respect to each gender along with the information
about whether they survived or not.

```python
sns.boxplot(data=df, x='sex', y='age', hue='survived', palette='husl')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

# Use violin plot and others plots from seaborn package

```python
sns.violinplot(data=df, x='sex', y='age', hue='survived', palette='husl')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.histplot(df["fare"],bins=20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Fare")
plt.show()
```
