# Use the Seaborn library to see if we can find any patterns in the data.

Pair Plot: Grid of scatterplots for pairwise relationships in the dataset. This can be helpful to visualize multiple pairwise relationships in one grid.

```python
sns.pairplot(df,hue='survived',palette='husl')
```

Count Plot: Similar to a bar plot but specifically for counting the occurrences of each category in a categorical variable.

```python
sns.countplot(data=df, x='class', hue='survived', palette='husl')
plt.show()
```

```python
sns.violinplot(data=df, x='sex', hue='survived', palette='husl')
plt.show()
```

```python
sns.barplot(data=df, x='sex',y='survived')
plt.show()
```

Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a
histogram.

```python
sns.histplot(df['fare'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
